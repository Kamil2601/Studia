```julia
using Pkg
Pkg.add("Plots")
using Plots
plotly()
setprecision(1024)
using Printf
```

```julia
function arctg_taylor(x,k)
    x = BigFloat(x)
    sum = 0
    for i = 0:k
        if i % 2 == 0
            sum += x^(2*i+1)/(2*i+1)
        else
            sum -= x^(2*i+1)/(2*i+1)
        end
    end
    return sum
end
f(x)=arctg_taylor(x,100)
```

```julia
plot([atan, f],xlimits=[-1.5,1.5],label=["atan" "Taylor"],ylimits=[-1,1],lw=1.5)
```

```julia
plot(f,BigFloat(-1.2):0.01:BigFloat(1.2),ylimits=[-1,1])
```

```julia
function avg_error(k)
    iterations = 0
    error_sum = BigFloat(0.0)
    for i = -0.99:0.01:0.99
        if i!=0
            x = arctg_taylor(BigFloat(i), k)
            y = atan(BigFloat(i))
            if y!=0
                error_sum += BigFloat(abs((x-y)/y))
                iterations+=1
            end
        end
    end
    error_avg = error_sum/BigFloat(iterations)
    return error_avg
end
```

```julia
using Printf
for k in [1,5,10,50,100,500,1000, 5000, 10000, 100000]
    t = avg_error(k)
    @printf "%0.10e\n" t
end
```

```julia
arcctg_taylor(x,k) = BigFloat(pi/2)-arctg_taylor(x,k)
```

```julia
arccot(x)=BigFloat(pi/2)-atan(BigFloat(x))
```

```julia
function avg_error(k,a,b)
    iterations = 0
    error_sum = BigFloat(0.0)
    for i = a:0.01:b
        if i!=0
            x = arctg_taylor(BigFloat(i), k)
            y = atan(BigFloat(i))
            if y!=0
                error_sum += BigFloat(abs((x-y)/y))
                iterations+=1
            end
        end
    end
    error_avg = error_sum/BigFloat(iterations)
    return error_avg
end
```

```julia
@printf "%0.10e\n" avg_error(1000,0.0,0.2)
@printf "%0.10e\n" avg_error(1000,0.2,0.4)
@printf "%0.10e\n" avg_error(1000,0.4,0.6)
@printf "%0.10e\n" avg_error(1000,0.6,0.8)
@printf "%0.10e\n" avg_error(1000,0.8,0.99)
@printf "%0.10e\n" avg_error(1000,0.0,0.70)
```

```julia
function sin_taylor(x,k)
    x = BigFloat(x)
    sum = 0
    power = x
    fact = 1
    for i = 0:k
        if i % 2 == 0
            sum += BigFloat(power/fact)
        else
            sum -= BigFloat(power/fact)
        end
        power = BigFloat(power*x^2)
        fact = BigFloat(fact*(2*(i+1))*(2*(i+1)+1))
        #@printf "%0.10e\n" power/fact
    end
    return sum
end
```

```julia
function cos_taylor(x,k)
    x = BigFloat(x)
    sum = 0
    power = 1
    fact = 1
    for i = 0:k
        if i % 2 == 0
            sum += BigFloat(power/fact)
        else
            sum -= BigFloat(power/fact)
        end
        power = BigFloat(power*x^2)
        fact = BigFloat(fact*(2*(i+1)-1)*(2*(i+1)))
        #@printf "%0.10e\n" power/fact
    end
    return sum
end
```

```julia
function tg_taylor(x,k)
    return sin_taylor(x,k)/cos_taylor(x,k)
end
```

```julia
tg(x)=tg_taylor(x,100)
```

```julia
function bisect(f,y,a,b, epsilon)
    p = BigFloat(a)
    q = BigFloat(b)
    
    while (abs(p-q)>epsilon)
        s = (p+q)/2.0
        if abs(f(s)-y) <= epsilon
            break
        elseif (f(s)-y) * (f(p)-y) < 0
            q = s 
        else
            p = s
        end
    end

    return (p+q)/2
end
```

```julia
arctg(x) = bisect(tg,x,-pi/2,pi/2,10^(-300))
```

```julia
accot(x)=pi/2-atan(x)
```

```julia
function ctg_taylor(x,k)
    if x == pi/2
        return 0
    else return cos_taylor(x,k)/sin_taylor(x,k)
    end
end
ctg(x)=ctg_taylor(x,100)
```

```julia
plot(ctg,xlimits=[0,pi],ylimits=[-5,5])
```

```julia
ctg(x) = ctg_taylor(x,100)
arcctg(x) = bisect(ctg,x,0,pi,10^(-16))
```

```julia
plot([accot,arcctg],xlimits=[-5,5],label=["pi/2-atan()" "bisekcja"],ylimits=[-5,5],lw=2)
```
